# Import

```python
import pandas as pd
```

# Data (President & Governor, v3.4)

```python
url_p = "https://raw.githubusercontent.com/sw1kwon/korean-elections/refs/heads/main/temp/v3/temp3_4_president.csv"
df_p = pd.read_csv(url_p)
df_p.head()
```

```python
url_g = "https://raw.githubusercontent.com/sw1kwon/korean-elections/refs/heads/main/temp/v3/temp3_4_governor.csv"
df_g = pd.read_csv(url_g)
df_g.head()
```

# Merge

```python
df_tidy = pd.concat([df_p, df_g], ignore_index=True)
df_tidy
```

```python
df_tidy.info()
```

# Preprocessing

```python
mapping_dictionary = {
    '대통령': '대선',
    '광역단체장': '지선'
}

df_tidy = df_tidy.replace({'선거종류': mapping_dictionary})

df_tidy
```

```python
df_tidy = (
    df_tidy
    .astype({'선거년도': 'object'})
    .loc[:, ['지역', '선거년도', '선거종류'] + [col for col in df_tidy.columns if col not in ['지역', '선거년도', '선거종류']]]
)

df_tidy.info()
```

```python
df_tidy
```

```python
df_tidy.to_csv("tidy1_combined.csv", index=False, encoding="utf-8-sig")
```
